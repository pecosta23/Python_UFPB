```python
def pagar_divida(valor_divida, pagamento_mensal):
    pagamento = 1

    while valor_divida > 0:
        valor_anterior = valor_divida
        valor_divida -= pagamento_mensal

        if valor_divida < 0:
            valor_divida = 0

        print(f'Pagamento {pagamento}: Antes = {valor_anterior}, Depois = {valor_divida}')

        pagamento += 1

valor_divida = int(input())
pagamento_mensal = int(input())

pagar_divida(valor_divida, pagamento_mensal)
```
